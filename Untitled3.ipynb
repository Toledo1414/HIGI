```python
import os
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
def Kepler(x0, y0, z0, vx0, vy0, vz0, t0=0, dt=0.01 ,nsteps=1000):
    rp = math.sqrt( x0**2 + y0**2 + z0**2)
    G = 4*(math.pi**2)
    x = [x0] 
    y = [y0]
    z = [z0]
    vx = [vx0]
    vy = [vy0]
    vz = [vz0]
    t = [t0]
    for i in range(1, nsteps):
        Vxi = vx0 - ( (G) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        Vyi = vy0 - ( (G) / rp**3 )*y0 *dt 
        yi = y0 + Vyi*dt
        Vzi = vz0 - ( (G) / rp**3 )*z0 *dt 
        zi = z0 + Vzi*dt
        x.append(xi)
        y.append(yi)
        z.append(zi)
        t.append(t0 + dt*i)
        x0 = xi
        y0 = yi
        z0 = zi
        vx0 = Vxi 
        vy0 = Vyi
        vz0 = Vzi
        
    return x, y, z, vx, vy, vz, t
```

```python
#mercurio
x,y,z,vx,vy,vz,t= Kepler(0.0508011, -0.4011642, -0.2195576, 0.02231875,0.00485553,0.00027924)
plt.plot(x,y)
```
